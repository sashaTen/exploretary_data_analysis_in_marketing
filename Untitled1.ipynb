```python
import pandas as pd
from matplotlib import pyplot as plt

# Read the CSV file
df = pd.read_csv('https://raw.githubusercontent.com/nailson/ifood-data-business-analyst-test/master/ml_project1_data.csv')

# Print the head of the data
print(df.head(15).to_string())
```

```python
print(df.tail(10).to_string())
```

```python
print(df.isnull().sum())
```

```python
median_income =   df['Income'].median()
df["Income"]=df["Income"].fillna(median_income)
print(df["Income"].isnull().sum())
```

```python

plt.hist(df["Income"], bins =  80)
plt.show()
```

```python
df.describe()
```

```python
std_of_income   =   df["Income"].std()*3
range=df["Income"].mean()+std_of_income
df.loc[df["Income"]>range, "Income"] =   df["Income"].median()
```

```python
print(df["Income"].describe())
plt.hist(df["Income"] ,  bins   =   60)
plt.show()
```

```python
# now  lets  define  the data  types
list_of_continus  =["Income"  ,  "MntFruits" ,  "MntWines" ,  "MntMeatProducts" , "MntFishProducts" , "MntSweetProducts","MntGoldProds" ]
list_of_descrete = ["Year_Birth" ,   "Kidhome",   "Teenhome"   ,  "NumDealsPurchases" ,  "NumWebPurchases" ,  "NumCatalogPurchases" ,  "NumStorePurchases" ,  "NumWebVisitsMonth"]
list_of_nominal=["Marital_Status","AcceptedCmp3" , "AcceptedCmp4", "AcceptedCmp5", "AcceptedCmp1", "AcceptedCmp2" ,   "Complain"    ,    "Response"]
list_of_ordinal=["Education" ,    "Dt_Customer" ,  "Recency"]

for   column   in   df.columns:
 plt.hist(df[column],  bins=50)
 plt.title(column)
 plt.show()
  
```

```python
for   i  in   list_of_descrete:
 plt.boxplot(df[i])
 plt.title(i)
 plt.xlabel("Data")
 plt.ylabel("Value")

# display the plot
 plt.show()
```

```python
import numpy as np
median = df["Year_Birth"].median()
Q1 = np.percentile(df["Year_Birth"], 25)
Q3 = np.percentile(df["Year_Birth"], 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
outliers = (df["Year_Birth"] < lower_bound) | (df["Year_Birth"] > upper_bound)
df.loc[outliers, "Year_Birth"] = median

print(df["Year_Birth"].min())
plt.boxplot(df["Year_Birth"])
plt.title("year of birth")
plt.xlabel("Data")
plt.ylabel("Value")

# display the plot
plt.show()
```

```python
print(df[["AcceptedCmp1" ,   "AcceptedCmp5", "AcceptedCmp4","Response"]].tail(10))
```

```python
df['Dependents'] = df['Kidhome'] + df['Teenhome']

# Year becoming a Customer
df['Year_Customer'] = pd.DatetimeIndex(df['Dt_Customer']).year

# Total Amount Spent
mnt_cols = [col for col in df.columns if 'Mnt' in col]
df['TotalMnt'] = df[mnt_cols].sum(axis=1)

# Total Purchases
purchases_cols = [col for col in df.columns if 'Purchases' in col]
df['TotalPurchases'] = df[purchases_cols].sum(axis=1)

# Total Campaigns Accepted
campaigns_cols = [col for col in df.columns if 'Cmp' in col] + ['Response'] # 'Response' is for the latest campaign
df['TotalCampaignsAcc'] = df[campaigns_cols].sum(axis=1)
```

```python
df[['ID', 'Dependents', 'Year_Customer', 'TotalMnt', 'TotalPurchases', 'TotalCampaignsAcc']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute pairwise correlations between features
corr = df.corr()

# Create heatmap of correlation matrix
fig, ax = plt.subplots(figsize=(15, 12))
sns.heatmap(corr, cmap='coolwarm', ax=ax)

# Set x-axis and y-axis labels
ax.set_xticklabels(corr.columns)
ax.set_yticklabels(corr.columns)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Display plot
plt.show()
```

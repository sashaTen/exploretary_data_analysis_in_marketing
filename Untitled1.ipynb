```python
import pandas as pd
from matplotlib import pyplot as plt

# Read the CSV file
df = pd.read_csv('https://raw.githubusercontent.com/nailson/ifood-data-business-analyst-test/master/ml_project1_data.csv')

# Print the head of the data
print(df.head(15).to_string())
```

```python
print(df.tail(10).to_string())
```

```python
print(df.isnull().sum())
```

```python
median_income =   df['Income'].median()
df["Income"]=df["Income"].fillna(median_income)
print(df["Income"].isnull().sum())
```

```python
print(df.isnull().sum())
```

```python

plt.hist(df["Income"], bins =  80)
plt.show()
```

```python
df.describe()
```

```python
std_of_income   =   df["Income"].std()*3
range=df["Income"].mean()+std_of_income
max   =   df["Income"][(df['Income'] > range)]
print(max)
```

```python

df.loc[df["Income"]>range, "Income"] =   df["Income"].median()
```

```python
print(df["Income"].describe())
plt.hist(df["Income"] ,  bins   =   60)
plt.show()
```
